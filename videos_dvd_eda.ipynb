### Sentiment Analysis

```python
videos_2015['label'] = videos_2015['star_rating'].apply(lambda x:0 if x < 4 else 1)
videos_2015.head()
```

```python
import nltk
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer
import string

def get_pos_tag(pos_tag):
    if pos_tag.startswith('J'):
        return wordnet.ADJ
    elif pos_tag.startswith('V'):
        return wordnet.VERB
    elif pos_tag.startswith('N'):
        return wordnet.NOUN
    elif pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def clean_text(text):
    # lower text
    text = text.lower()
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # remove words that contain numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    # remove stop words
    stop = stopwords.words('english')
    text = [x for x in text if x not in stop]
    # remove empty tokens
    text = [t for t in text if len(t) > 0]
    # pos tag text
    pos_tags = pos_tag(text)
    # lemmatize text
    text = [WordNetLemmatizer().lemmatize(t[0], get_pos_tag(t[1])) for t in pos_tags]
    # remove words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
    text = " ".join(text)
    return(text)

# videos_2015['clean_review'] = videos_2015['review_headline'].apply(lambda x: clean_text(x))
# videos_2015.head()

# clean text data
videos_2015_sample = videos_2015.sample(frac=0.1, replace=False, random_state=42)
videos_2015_sample['clean_review'] = videos_2015_sample['review_body'].apply(lambda x: clean_text(x))
videos_2015_sample.head()
```

```python
# add sentiment anaylsis columns
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()
videos_2015_sample['sentiments'] = videos_2015_sample['clean_review'].apply(lambda x: sid.polarity_scores(x))
videos_2015_sample = pd.concat([videos_2015_sample.drop(['sentiments'], axis=1), videos_2015_sample['sentiments'].apply(pd.Series)], axis=1)
videos_2015_sample.head(20)
```

```python
# add number of characters column
videos_2015_sample["nb_chars"] = videos_2015_sample['clean_review'].apply(lambda x: len(x))

# add number of words column
videos_2015_sample["nb_words"] = videos_2015_sample['clean_review'].apply(lambda x: len(x.split(" ")))
```
