```python
import os
import json
import csv
import gzip
import pandas as pd
```

## Physical

### Cleaning

```python
path = '/Users/auraulloa/Desktop/DataSciencePipeline/'
physical_path = path + 'CDs_and_Vinyl.json.gz'
```

```python
phys_reviews = []
with gzip.open(physical_path) as f:
    for l in f:
        phys_reviews.append(json.loads(l.strip()))
```

```python
len(phys_reviews)
```

```python
phys_review_df = pd.DataFrame.from_dict(phys_reviews)
```

```python
len(phys_review_df)
```

```python
phys_review_df.head()
```

```python
phys_review_df = phys_review_df.drop("reviewerName", 1)
phys_review_df = phys_review_df.drop("reviewerID", 1)
phys_review_df = phys_review_df.drop("image", 1)
```

```python
phys_review_df = phys_review_df.drop("unixReviewTime", 1)
```

```python
phys_review_df["style"] = phys_review_df.apply(lambda x: x["style"]["Format:"] if type(x["style"]) == dict and "Format:" in x["style"] else x["style"], axis=1)
```

```python
phys_review_df[phys_review_df["style"].notna()]["style"].astype(str).unique()
```

```python
phys_review_df = phys_review_df.drop(4510416)
```

```python
phys_review_df["style"] = phys_review_df.apply(lambda x: ' Audio CD' if x["style"] == {'Size:': ' Small'} else x["style"], axis = 1)
```

```python
phys_review_df = phys_review_df[phys_review_df["style"] != {'Product Packaging:': ' Standard Packaging'}]
```

```python
phys_review_df['vote'] = phys_review_df.apply(lambda x: '0' if type(x['vote']) != str else x['vote'], axis = 1)
```

```python
phys_review_df["style"] = phys_review_df.apply(lambda x: x["style"][1:] if type(x["style"]) == str else x["style"], axis=1)
```

```python
phys_review_df[phys_review_df["style"].notna()]["style"].astype(str).unique()
```

```python
phys_review_df.head()
```

### Merge with meta data

```python
products = pd.read_csv("physicals.csv") 
```

```python
products.head()
```

```python
final_merge = products.merge(phys_review_df, how="inner", on="asin")
```

```python
len(final_merge)
```

```python
final_merge.head()
```

```python
final_merge.to_csv('final-physical.csv', index = False, header=True)
```

## Digital

### Cleaning

```python
digital_path = path + 'Digital_Music.json.gz'
```

```python
dig_reviews = []
with gzip.open(digital_path) as f:
    for l in f:
        dig_reviews.append(json.loads(l.strip()))
```

```python
len(dig_reviews)
```

```python
dig_review_df = pd.DataFrame.from_dict(dig_reviews)
```

```python
len(dig_review_df)
```

```python
dig_review_df.head()
```

```python
dig_review_df.tail()
```

```python
dig_review_df = dig_review_df.drop("reviewerID", 1)
dig_review_df = dig_review_df.drop("reviewerName", 1)
dig_review_df = dig_review_df.drop("unixReviewTime", 1)
dig_review_df = dig_review_df.drop("image", 1)
```

```python
dig_review_df.head()
```

```python
dig_review_df["vote"] = dig_review_df.apply(lambda x: '0' if type(x["vote"]) == float else x["vote"], axis=1)
```

```python
dig_review_df["style"] = dig_review_df.apply(lambda x: x["style"]['Format:'] if type(x["style"]) == dict and 'Format:' in x["style"] else x["style"], axis=1)
```

```python
dig_review_df[dig_review_df["style"].notna()]["style"].astype(str).unique()
```

```python
dig_review_df = dig_review_df[dig_review_df["style"] != {'Color:': ' Black'}]
dig_review_df = dig_review_df[dig_review_df["style"] != {'Size:': ' O/S', 'Color:': ' Multicolor'}]
```

```python
dig_review_df.head()
```

```python
dig_review_df.tail()
```

```python
len(dig_review_df)
```

### Merge with Metadata

```python
dig_products = pd.read_csv("digitals.csv") 
```

```python
dig_products.head()
```

```python
final_dig_merge = dig_products.merge(dig_review_df, how="inner", on="asin")
```

```python
len(final_dig_merge)
```

```python
final_dig_merge.to_csv('final-digital.csv', index = False, header=True)
```
