```python
issa_match_videos = videos_2015[videos_2015['product_title'].isin(dvd_2015['product_title']) == True]
issa_match_videos
```

```python
titles = issa_match_videos['product_title'].unique()
titles
```

```python
import numpy as np

videos_dvd_2015 = pd.DataFrame(titles,columns=['titles'])
videos_2015_mean = videos_2015[videos_2015['product_title'].isin(titles)==True].groupby(['product_title']).mean()
videos_dvd_2015['star_rating_videos'] = np.array(videos_2015_mean['star_rating'])

videos_dvd_2015
```

```python
dvd_2015_mean = dvd_2015[dvd_2015['product_title'].isin(titles)==True].groupby(['product_title']).mean()
videos_dvd_2015['star_rating_dvd'] = np.array(dvd_2015_mean['star_rating'])

videos_dvd_2015
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [7, 7]

videos_dvd_2015.plot(kind='scatter', x='star_rating_videos', y='star_rating_dvd', xlabel='video star rating', ylabel='dvd star rating')
plt.title("Scatter plot showing video star ratings vs DVD star rating")
plt.show()
```
